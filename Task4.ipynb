```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Development_sample_4.csv')
df
```

```python
cross1=pd.crosstab(df['PRE_3'],df['FLAGBINARY'],margins=True,margins_name='Total')
cross2=pd.crosstab(df['PRE_3'],df['FLAGBINARY'],normalize='index')
```

```python
cross1
```

```python
cross1['CBads']=cross1.iloc[:,0].cumsum()
cross1['CGoods']=cross1.iloc[:,1].cumsum()
```

```python
cross1['PBads']=cross1['CBads']/747
cross1['PGoods']=cross1['CGoods']/4253
cross1['ABSDifference']=abs(cross1['PBads']-cross1['PGoods'])
```

```python
cross1
```

```python
max(cross1['ABSDifference'])
```

```python
cross1['CTotal']=cross1['Total'].cumsum()
cross1['AtAbove']=5000-cross1['CTotal']+cross1['Total']
cross1['PAtAbove']=cross1['AtAbove']/5000
cross1['AtAboveBads']=747-cross1['CBads']
cross1['AtAboveGoods']=4253-cross1['CGoods']
cross1['CBadRate']=cross1['AtAboveBads']/cross1['AtAbove']
```
